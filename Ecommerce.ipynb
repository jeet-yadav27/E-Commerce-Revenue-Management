<font color='blue'>
    <font size = 5>
    In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business.
    The E-Commerce company has the following rule for computing their own revenue: </font>
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font>

##  the Total Revenue generated by the E-Commerce company over all orders placed.

```python
#Function to calculate the revenue of the company as per the condition provided
def Rev(row):
      if row["discounted_price"]>600:
       val=row["discounted_price"]*.25
      elif ((row["discounted_price"]>350) & (row["discounted_price"]<=600)):
       val=row["discounted_price"]*.15
      elif ((row["discounted_price"]>100) & (row["discounted_price"]<=350)):
       val=row["discounted_price"]*.1
      else:
       val=row["discounted_price"]*.05
      return val
```

```python
#to call the Function and add the value of each particular products in another column name as Ecom_Revenue
df['Ecom_Revenue']= df.apply(Rev, axis=1)
```

```python
# To Calculate the Total Revenue of the E-commerce Company
Total_Revenue=df["Ecom_Revenue"].sum()
```

```python
# display the Total Revenue
print("The Revenue of the E-Commerce Company is: Rs.",Total_Revenue, sep="")
```
